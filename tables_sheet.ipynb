```python
import psycopg2
import pandas as pd

conn = psycopg2.connect(
    host="localhost",
    port=5432,
    database="trucking_ops",
    user="postgres",
    password="yournewpassword"
)

# List all tables
pd.read_sql_query("""
    SELECT table_name
    FROM information_schema.tables
    WHERE table_schema = 'public'
    ORDER BY table_name;
""", conn)
```

```python
tables = pd.read_sql_query("""
    SELECT table_name
    FROM information_schema.tables
    WHERE table_schema = 'public'
    ORDER BY table_name;
""", conn)
```

```python
for table in tables['table_name']:
    print(f"\n--- {table} ---")
    df = pd.read_sql_query(f"SELECT * FROM {table} LIMIT 3;", conn)
    display(df)
```

```python
# Get all tables with row counts and column names
tables = pd.read_sql_query("""
    SELECT table_name
    FROM information_schema.tables
    WHERE table_schema = 'public'
    ORDER BY table_name;
""", conn)

for table in tables['table_name']:
    cols = pd.read_sql_query(f"""
        SELECT column_name, data_type
        FROM information_schema.columns
        WHERE table_name = '{table}'
        ORDER BY ordinal_position;
    """, conn)
    count = pd.read_sql_query(f"SELECT COUNT(*) as rows FROM {table};", conn)
    print(f"\n--- {table} ({count['rows'][0]:,} rows) ---")
    display(cols)
```
